```python
import pandas as pd
from collections import defaultdict
import json
import difflib
```

```python
def get_equal_rate(str1, str2):
    return difflib.SequenceMatcher(None, str1, str2).quick_ratio()
```

```python
filename="TIME.xlsx"
data = pd.read_excel(filename, sheet_name = "Sheet1")

name_team_dict={}
with open('teams2.json', 'r') as f:
    teams_dict = json.load(f)

threshold=0.6
    
     
for i,item in enumerate(data["ç”¨æˆ·æ˜µç§°"]):
    
    get_teamname = name_team_dict.get(item,0)
    if get_teamname!=0:
        teams_dict[get_teamname]["çº¿ä¸Šè‡ªä¹ æ€»æ—¶é•¿"]+=data['ä¸“æ³¨æ—¶é•¿'][i]
        continue
        
    best_score=0
    team_name_record=0
    match_name_record=0
    
    for team_name,team_info in teams_dict.items():
        for bk_name in team_info["åŒ¹é…å¤‡é€‰åç§°"]:
            
            if len(item)>4 and len(bk_name)<4:
                score=get_equal_rate(item[0:3],bk_name) 
            elif len(item)<4 and len(bk_name)>4:
                score=get_equal_rate(item,bk_name[0:3])                
            else:
                score=get_equal_rate(item,bk_name)     
            
            if score>best_score and score>threshold:
                best_score=score
                team_name_record=team_name   
                match_name_record=bk_name

#     å¯ä¿¡åº¦åœ¨0.7ä»¥ä¸‹ï¼Œå»ºè®®äººå·¥æ ¸æŸ¥
#     if best_score <0.7 and best_score>0:   
#         print("ã€",item,"ã€‘ã€",match_name_record,"ã€‘",best_score)
    
    if best_score>threshold:
        print("ã€",item,"ã€‘ã€",match_name_record,"ã€‘",best_score)
        name_team_dict[item]=team_name_record
        teams_dict[team_name_record]["çº¿ä¸Šè‡ªä¹ æ€»æ—¶é•¿"]+=data['ä¸“æ³¨æ—¶é•¿'][i]    
        
```
